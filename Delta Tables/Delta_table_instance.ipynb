### Delta Instance

```
#importing modules
from pyspark.sql import SparkSession
from delta.tables import*
```

```
DT = DeltaTable.create(spark) \
    .tableName('Employe_demo3') \
    .addColumn('emp_id','INT') \
    .addColumn('emp_name','STRING') \
    .addColumn('gender','STRING') \
    .addColumn('salary','INT') \
    .addColumn('dept','STRING') \
    .property('description','table created for demo purpose') \
    .location('/FileStore/table/delta/createtable3') \
    .execute()
```

```
%sql
insert into employe_demo3 values(100,'shiva','M',2000,'IT');
insert into employe_demo3 values(200,'sai','M',8000,'HR');
insert into employe_demo3 values(300,'charan','M',6000,'sales');
insert into employe_demo3 values(400,'niranjan','M',7000,'DE');
insert into employe_demo3 values(500,'setha','F',8000,'BI')
```
